## <font color='green'>Actividad 2</font>

Supongamos que una tienda en l√≠nea que ofrece env√≠o a domicilio basado en una tabla de descuentos seg√∫n el monto del pedido.

- Si el pedido es mayor o igual a 1500, se realiza un env√≠o express gratis.
- Si es mayor o igual a 1000 pero menor a 1500, el env√≠o es gratis pero se realiza mediante el servicio est√°ndar;.
- Si es mayor o igual a 500 pero menor a 1000, el env√≠o tiene un costo de 10 y se realiza mediante el servicio est√°ndar.
- Si el pedido es menor a 500, no se puede procesar.

### <font color='green'>Soluci√≥n</font>

```python
# Puedes modificar este valor. ¬°O podr√≠a solicitarse con input!
monto_pedido = 1000
```

```python
if monto_pedido >= 1500:
    print("üéâ Tu pedido califica para env√≠o express gratis üöö")

elif monto_pedido >= 1000 and monto_pedido < 1500:
    print("üéâ Tu pedido califica para env√≠o est√°ndar gratis üöõ")

elif monto_pedido >= 500 and monto_pedido < 1000:
    print("üöõ Tu pedido ser√° enviado por env√≠o est√°ndar por un costo de $10")

else:
    print("üôÅ Lo sentimos, no podemos procesar tu pedido ya que el monto m√≠nimo de pedido es de $500")
```
