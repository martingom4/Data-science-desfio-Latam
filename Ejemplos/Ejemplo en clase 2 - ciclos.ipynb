 # <font color='green'>Ciclo for</font>

```python
frutas = ["manzana", "banana", "kiwi", "naranja"]

for fruta in frutas:
    print(fruta)
```

Tambi√©n podemos utilizar el **ciclo for** para iterar sobre una cadena de caracteres:

```python
mensaje = "Hola, mundo!"

for caracter in mensaje:
    print(caracter)
```

Los diccionarios tienen una forma especial de recorrerse utilizando un ciclo for. Observa el siguiente bloque.

```python
refranes = {"Perro que ladra": "no muerde",
            "A quien madruga": "Dios le ayuda",
            "No dejes para ma√±ana": "lo que puedas hacer hoy"}

for llave, valor in refranes.items():
    refran_completo = llave + ", " + valor
    print(refran_completo)
```

 # <font color='green'>Ciclo while</font>

```python
numero = 1

while numero <= 5:
    print(numero)
    numero = numero + 1
```

Tambi√©n podemos utilizar el **ciclo while** para repetir un bloque de c√≥digo hasta que el usuario ingrese un valor espec√≠fico. Por ejemplo:

```python
respuesta = ""

while respuesta != "s":
    respuesta = input("¬øQuieres finalizar? (s/n)")

print("Fin del programa.")
```

**Explicaci√≥n:** Este c√≥digo le pedir√° al usuario que ingrese "s" o "n" para decidir si quiere finalizar o no. Mientras el usuario no ingrese "s", el ciclo se seguir√° repitiendo. Cuando el usuario finalmente ingrese "s", el ciclo se detendr√° y el programa imprimir√° "Fin del programa.".

```python
opcion = 0

while opcion != 3:
    print("¬øQu√© acci√≥n deseas realizar?")
    print("1. Saludar")
    print("2. Sumar dos n√∫meros")
    print("3. Salir")

    opcion = int(input("Ingrese su opci√≥n: "))

    if opcion == 1:
        print("¬°Hola, mundo!")
    elif opcion == 2:
        num1, num2 = map(int, input("Ingrese dos n√∫meros separados por espacio: ").split())
        print("El resultado de la suma es:", num1 + num2)
    elif opcion == 3:
        print("Adi√≥s.")
    else:
        print("Opci√≥n inv√°lida.")
```

**Explicaci√≥n:** En este ejemplo, el **ciclo while** se ejecuta hasta que el usuario seleccione la opci√≥n "3" para salir. Si el usuario selecciona la opci√≥n "1", el programa imprime "¬°Hola, mundo!". Si el usuario selecciona la opci√≥n "2", el programa le pide al usuario que ingrese dos n√∫meros separados por espacio y luego imprime el resultado de la suma. Si el usuario selecciona una opci√≥n inv√°lida, el programa le informa al usuario que su opci√≥n es inv√°lida y vuelve a mostrar el men√∫.




 # <font color='green'>Ciclos anidados</font>

```python
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
suma = 0

for fila in matriz:
    for elemento in fila:
        suma += elemento
        print(suma)
```

Podemos observar que primero recorremos cada elemento de la matriz (que, a su vez, son listas). Para cada uno de estos elementos, hacemos a su vez un recorrido por ellos.

 # <font color='green'>Break</font>
El comando **break** nos permite terminar un ciclo verificando una condici√≥n, antes de que se complete.

Por ejemplo: queremos que el programa le solicite una contrase√±a al usuario, con un m√°ximo de 3 intentos (luego de los cuales se bloquea). Podemos implementar esto de la siguiente manera:

```python
password="adivina"
intentos = 0

while intentos < 3:
    respuesta = input("Ingrese contrase√±a: ")
    intentos+=1
    restantes=3-intentos
    if respuesta != password and restantes !=0:
        print(f"Contrase√±a incorrecta. Le quedan {restantes} intentos")
    else:
        break

if respuesta == password:
    print(f"¬°Acceso permitido!")
else:
    print(f"Acceso denegado")
```

```python
import random
a=random.randint(1,20)
intentos = 0

while intentos < 5:
    respuesta = int(input("Ingresa un n√∫mero: "))
    intentos+=1
    if respuesta < a:
        print("El n√∫mero es mayor")
    if respuesta > a:
        print("El n√∫mero es menor")
    if respuesta == a:
        break

if respuesta == a:
    print(f"¬°Has adivinado!")
else:
    print(f"¬°Has fallado! El n√∫mero es {a}")
```
