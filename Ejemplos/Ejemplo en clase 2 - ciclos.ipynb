 # <font color='green'>Ciclo for</font>

```python
frutas = ["manzana", "banana", "kiwi", "naranja"]

for fruta in frutas:
    print(fruta)
```

También podemos utilizar el **ciclo for** para iterar sobre una cadena de caracteres:

```python
mensaje = "Hola, mundo!"

for caracter in mensaje:
    print(caracter)
```

Los diccionarios tienen una forma especial de recorrerse utilizando un ciclo for. Observa el siguiente bloque.

```python
refranes = {"Perro que ladra": "no muerde",
            "A quien madruga": "Dios le ayuda",
            "No dejes para mañana": "lo que puedas hacer hoy"}

for llave, valor in refranes.items():
    refran_completo = llave + ", " + valor
    print(refran_completo)
```

 # <font color='green'>Ciclo while</font>

```python
numero = 1

while numero <= 5:
    print(numero)
    numero = numero + 1
```

También podemos utilizar el **ciclo while** para repetir un bloque de código hasta que el usuario ingrese un valor específico. Por ejemplo:

```python
respuesta = ""

while respuesta != "s":
    respuesta = input("¿Quieres finalizar? (s/n)")

print("Fin del programa.")
```

**Explicación:** Este código le pedirá al usuario que ingrese "s" o "n" para decidir si quiere finalizar o no. Mientras el usuario no ingrese "s", el ciclo se seguirá repitiendo. Cuando el usuario finalmente ingrese "s", el ciclo se detendrá y el programa imprimirá "Fin del programa.".

```python
opcion = 0

while opcion != 3:
    print("¿Qué acción deseas realizar?")
    print("1. Saludar")
    print("2. Sumar dos números")
    print("3. Salir")

    opcion = int(input("Ingrese su opción: "))

    if opcion == 1:
        print("¡Hola, mundo!")
    elif opcion == 2:
        num1, num2 = map(int, input("Ingrese dos números separados por espacio: ").split())
        print("El resultado de la suma es:", num1 + num2)
    elif opcion == 3:
        print("Adiós.")
    else:
        print("Opción inválida.")
```

**Explicación:** En este ejemplo, el **ciclo while** se ejecuta hasta que el usuario seleccione la opción "3" para salir. Si el usuario selecciona la opción "1", el programa imprime "¡Hola, mundo!". Si el usuario selecciona la opción "2", el programa le pide al usuario que ingrese dos números separados por espacio y luego imprime el resultado de la suma. Si el usuario selecciona una opción inválida, el programa le informa al usuario que su opción es inválida y vuelve a mostrar el menú.




 # <font color='green'>Ciclos anidados</font>

```python
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
suma = 0

for fila in matriz:
    for elemento in fila:
        suma += elemento
        print(suma)
```

Podemos observar que primero recorremos cada elemento de la matriz (que, a su vez, son listas). Para cada uno de estos elementos, hacemos a su vez un recorrido por ellos.

 # <font color='green'>Break</font>
El comando **break** nos permite terminar un ciclo verificando una condición, antes de que se complete.

Por ejemplo: queremos que el programa le solicite una contraseña al usuario, con un máximo de 3 intentos (luego de los cuales se bloquea). Podemos implementar esto de la siguiente manera:

```python
password="adivina"
intentos = 0

while intentos < 3:
    respuesta = input("Ingrese contraseña: ")
    intentos+=1
    restantes=3-intentos
    if respuesta != password and restantes !=0:
        print(f"Contraseña incorrecta. Le quedan {restantes} intentos")
    else:
        break

if respuesta == password:
    print(f"¡Acceso permitido!")
else:
    print(f"Acceso denegado")
```

 # <center><font color='green'>Soluciones</font></center>

```python
# Pedir al usuario que ingrese su edad
edad = int(input("¿Cuántos años tienes? "))

# Verificar la edad y mostrar un mensaje correspondiente
if edad < 18:
    print("Eres menor de edad")
elif edad >= 18 and edad < 65:
    print("Eres mayor de edad")
else:
    print("Eres un adulto mayor")
```

```python
num1 = int(input("Introduce el primer número: "))
num2 = int(input("Introduce el segundo número: "))

if num1 > num2:
    print("El primer número es mayor")
elif num2 > num1:
    print("El segundo número es mayor")
else:
    print("Los números son iguales")
```

```python
import random
a=random.randint(1,20)
intentos = 0

while intentos < 5:
    respuesta = int(input("Ingresa un número: "))
    intentos+=1
    if respuesta < a:
        print("El número es mayor")
    if respuesta > a:
        print("El número es menor")
    if respuesta == a:
        break

if respuesta == a:
    print(f"¡Has adivinado!")
else:
    print(f"¡Has fallado! El número es {a}")
```
