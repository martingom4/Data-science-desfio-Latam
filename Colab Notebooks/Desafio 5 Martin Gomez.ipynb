# 1.Carga los datos y crea un DataFrame con cada uno de ellos.

```python
#importar librerias necesarias para hacer el desafio
import pandas as pd
import numpy as np

df_officers = pd.read_pickle('/content/drive/MyDrive/Cosas importantes/Curso Desafio/Python/Curso de python/Apoyodesafio/officers.pkl')
df_incidents = pd.read_pickle('/content/drive/MyDrive/Cosas importantes/Curso Desafio/Python/Curso de python/Apoyodesafio/incidents.pkl')
df_subjects = pd.read_pickle('/content/drive/MyDrive/Cosas importantes/Curso Desafio/Python/Curso de python/Apoyodesafio/subjects.pkl')
```

# 2.Genera una tabla que contenga la unión de las 3 tablas

```python
df = df_incidents.merge(df_officers, how="left", on="case_number").merge(df_subjects, how="left", on="case_number", suffixes=('_officer', 'subject'))

df.head()
```

#  3.Verifica si hay filas duplicadas; si es así, elimínalas.

```python
print(len(df_incidents))
print(len(df))
```

```python
df.drop_duplicates(inplace=True)
```

# 4.¿Cuántos sujetos de género F hay en el DataFrame resultante?

```python
# Contar sujetos de género F
Total_mujeres = df['gendersubject'].value_counts().get('F', 0)
print(f"Numero de mujeres totales: {Total_mujeres}")

```

# 5.¿Encuántos números de caso hay por lo menos una sospechosa que sea mujer?

```python
# Filtrar casos con al menos una sospechosa mujer
Casos_femeninos = df[df['gendersubject'] == 'F']['case_number'].unique()
num_Casos_femeninos = len(Casos_femeninos)
print(f"números de caso hay por lo menos una sospechosa que sea mujer: {num_Casos_femeninos }")
```

## 6.Genera una tabla pivote que muestre en las filas el género del oficial y en las columnas el género del subject. ¿Cómo interpretas los valores que muestra esta vista?

```python
# Crear tabla pivote con el género del oficial y del sujeto
pivot_table = pd.pivot_table(df, values='case_number', index='gender_officer', columns='gendersubject', aggfunc='count')

# Mostrar la tabla pivote
pivot_table

```

*   Hay significativamente menos casos con oficiales femeninas.

### 3. Utiliza la siguiente lista de valores que serán considerados como nulos: ["na", "NA",-1, "0", "-1", "null", "n/a", "N/A", "NULL"]

```python
# Calcular la cantidad de valores nulos en cada columna del DataFrame
df.isnull().sum()
```

```python
#identicar el numero de valores nulos
valores_nulos = ["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"]

df.isin(valores_nulos).sum()
```

```python
# Reemplazar los valores considerados nulos con np.nan
df.replace(valores_nulos, np.nan, inplace=True)
df.isnull().sum()
```

## 4.Elimina todas las filas con datos faltantes.

```python
# Eliminar filas con datos faltantes
df.dropna(inplace=True)
```

## 5.Apartir de la columna “Salary Estimate”, genera dos columnas: Salario Estimado Mínimo y Máximo.

```python
# Ver los salarios estimados que estan en el documento
df[['Salary Estimate']].head()
```

```python
df['Min Estimated Salary'] = df['Salary Estimate'].apply(lambda x: x.split('-')[0])
df['Max Estimated Salary'] = df['Salary Estimate'].apply(lambda x: x.split('-')[1])

# Convertir las columnas de salario a números, manejando posibles errores para poder hacer la tabla pivote
df['Min Estimated Salary'] = pd.to_numeric(df['Min Estimated Salary'], errors='coerce')
df['Max Estimated Salary'] = pd.to_numeric(df['Max Estimated Salary'], errors='coerce')

df[['Salary Estimate','Min Estimated Salary','Max Estimated Salary']].head()
```

## *6*.Realiza la recodificación de la columna Size con los valores de la siguiente tabla:

```python
# Recodificación de la columna Size
size_mapping = {
    "10000+ employees": "Mega Empresas",
    "5001 to 10000 employees": "Grandes Empresas",
    "1001 to 5000 employees": "Medianas Empresas",
    "201 to 500 employees": "Pequeñas Empresas",
    "51 to 200 employees": "Pequeñas Grandes Empresas",
    "501 to 1000 employees": "Microempresas",
    "Unknown": "Empresas sin Información"
}
df['Size'] = df['Size'].replace(size_mapping)
```

## 7.Finalmente, genera una tabla pivote que muestre la media del salario estimado mínimo y la media del salario estimado máximo por tamaño de empresa

```python
# Crear la tabla pivote
pivot_table_salary = pd.pivot_table(df, values=['Min Estimated Salary', 'Max Estimated Salary'], index='Size', aggfunc='mean')

# Mostrar la tabla pivote
pivot_table_salary
```
