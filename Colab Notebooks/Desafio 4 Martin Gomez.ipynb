```python
# se importa la librerias necesarias para este desafio
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Cosas importantes/Curso Desafio/Python/Curso de python/Apoyodesafio/resumen_resultados_astronautas.csv')
# se muestra lo que hay en el data frame y se ysa head para mostrar los primeros
df.head()
```

```python
#se agrega una columna de alturae ne metros
df['altura en mts'] = df['altura']/100
#se muestran los datos resultantes

df.head()
```

```python
#se calcula el IMC con la formula proporcionada de IMC Peso  / altura
df['IMC'] = df['peso']/(df['altura en mts']**2)
df.head()
```

```python
#se saca el prmedio y se le asigna el nombre, se suma las cuatro evaluaciones y se divide entre 4 que son la cantidad
df ['Promedio de evaluaciones'] = (df['evaluacion_fisica'] + df['evaluacion_psicologica'] + df['evaluacion_adaptativa']+ df['evaluacion_medica'])/4
df
```

```python
#se filtra el data frame por el IMC y que los valores este
df[(df['IMC']>=18.5) & (df['IMC']<=24.9)]
df
```

```python
#se hace el filtrado de los datos tanto el IMC y el promedio de evaluacion
df_resultado = df[(df['IMC']>=18.5) & (df['IMC']<=24.9) & (df['Promedio de evaluaciones'] >= 87)]
#se imprime el data frame resultante
df_resultado
```

```python
#se reinicia la columna de los indices para que empiece desde 0 hasta la cantidad de datos
df_resultado.reset_index(drop=True, inplace=True )
#se imprimen los resultados
df_resultado
```

```python
#Se cambia la calificacion a si segun los datos filtrados y pasaron las pruebas
df_resultado ['califica'] = 'Si'
#Se imprimen los resultados
df_resultado
```

```python
#se guarda el data frame de los resultados en separados por comas
df_resultado.to_csv('/content/drive/MyDrive/Cosas importantes/Curso Desafio/Python/Curso de python/Apoyodesafio/resumen_resultados_astronautas.csv', sep=',', index=False)
#se lee el csv para ver los resultados del data frame
pd.read_csv('/content/drive/MyDrive/Cosas importantes/Curso Desafio/Python/Curso de python/Apoyodesafio/resumen_resultados_astronautas.csv')
```
